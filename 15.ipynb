```python
import tensorflow
```

```python
from tensorflow.keras.layers import Dense, Input
```

```python
import nltk
```

```python
import re
```

```python
with open('./train') as file:
    data = file.readlines()
```

```python
for i in range(len(data)):
    data[i] = data[i].replace('.\n','')
    data[i] = data[i].lower()
    data[i] = re.sub('[^A-Za-z ]','',data[i])
```

```python
data
```

```python
tokens = []
for line in data:
    lst = nltk.word_tokenize(line)
    if len(lst) >= 5:
        tokens.append(lst)
```

```python
enc = {}
dec = {}
tokens
```

```python
code = 0
for line in tokens:
    for word in line:
        if word not in enc.keys():
            enc[word] = code
            code += 1
```

```python
enc
```

```python
code = 0
for i in enc:
    dec[code] = i
    code += 1
```

```python
dec
```

```python
context = []
target = []
```

```python
for sentence in tokens:
    for i in range(2,len(sentence)-2):
        context.append([ enc[sentence[i-2]], enc[sentence[i-1]], enc[sentence[i+1]], enc[sentence[i+2]] ])
        target.append(enc[sentence[i]])
```

```python
for i in range(len(target)):
    print(context[i],' :: ',target[i])
```

```python
model = tensorflow.keras.models.Sequential()
```

```python
model.add(Input(shape=(4)))
```

```python
model.add(Dense(64,activation='relu'))
model.add(Dense(512,activation='relu'))
model.add(Dense(1024,activation='relu'))
model.add(Dense(361,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
hist = model.fit(x=context, y=target, batch_size=32, epochs=200)
```
